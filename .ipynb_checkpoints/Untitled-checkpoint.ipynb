```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import datetime as dt

%matplotlib inline
```

```python
data = 'Hotel Booking Dataset.xlsx'
data
```

```python
df_data = pd.read_excel(data)
df_data
```

```python
df_data.loc[:, ['arrival_date_day_of_month', 'arrival_date_month', "arrival_date_year"]]
```

```python
f"{df_data['arrival_date_month'][0]}-" + \
f"{df_data['arrival_date_day_of_month'][0]}-" +\
f"{df_data['arrival_date_year'][0]}"
```

```python
def create_date_arrival(dataframe, record):
    
    arrival_date = f"{dataframe['arrival_date_year'][record]}-" + \
    f"{dataframe['arrival_date_month'][record]}-" +\
    f"{dataframe['arrival_date_day_of_month'][record]}"
    
    # arrival_date = dt.datetime.strptime('2011-03-07','%Y-%m-%d')
    
    return arrival_date
```

```python
for idx in df_data.index:
    df_data.loc[idx, "date_arrival"] = create_date_arrival(df_data, idx)
```

```python
df_data["date_arrival"] = pd.to_datetime(df_data["date_arrival"], yearfirst = True)
df_data["date_arrival"]
```

```python
df_data["reservation_status_date"] = \
        pd.to_datetime(df_data["reservation_status_date"], yearfirst = True)

df_data["reservation_status_date"]
```
